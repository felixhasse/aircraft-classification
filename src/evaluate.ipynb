```python
from datasets import FGVCDataset
from torch.utils.data import DataLoader
from torchvision import transforms
from models import EfficientNet_V2_S
from torchvision.models import EfficientNet_V2_S_Weights
import torch
```

```python
model_path = "model2.pth"
```

```python
transform = EfficientNet_V2_S_Weights.IMAGENET1K_V1.transforms()

model = EfficientNet_V2_S(70)
model.load_state_dict(torch.load(model_path))
model.eval()

dataset = FGVCDataset("test", "family", transform=transform)
dataloader = DataLoader(dataset, batch_size=1, shuffle=True)

image, label = next(iter(dataloader))

true_class = dataset.getClassFromLabel(label[0])
output = model(image)
softmax_output = torch.nn.functional.softmax(output, dim=1).squeeze()
class_probabilities = dataset.getLabeledClasses(softmax_output)
predicted_class = dataset.getClassFromLabel(softmax_output)
class_probabilities.sort(key=lambda x: x[1], reverse=True)

print(torch.argmax(softmax_output))

print(f"True class: {true_class}")
print(f"Predicted class: {predicted_class}")
print(f"Class probabilities: {class_probabilities}")

transforms.ToPILImage()(image[0]).show()



```
